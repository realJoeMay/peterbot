# Sites DB

### Create Site

```python
from peterbot.data.collections import sites

site = {
    "site_id": sites.get_max_site_id() + 1,
    "name": "Diocese of Fort Wayne-South Bend",
    "slug": "diocese-of-fort-wayne-south-bend",
    "site_role": "owned",
    "primary_purpose": "diocese",
    "site_type": "website",

    "url": "https://diocesefwsb.org",
    "domain": "diocesefwsb.org",

    "status": "active",
    "verified": False,

    "parishes_page": "https://diocesefwsb.org/find/"
  }

doc_id = sites.create_site(site)
print(f"Created site with id {doc_id}")
```

### Read Sites

```python
sites.get_site(doc_id)
```

### Delete Sites

```python
# by doc id
status = sites.delete_site(doc_id)
print(f"Deleted site with id {doc_id}: {status}")
```

```python
# by site id
site_id = 1
status = sites.delete_site_by_id(site_id)
print(f"Deleted site with site id {site_id}: {status}")
```

```python
# delete all sites
for i in range(1, sites.get_max_site_id() + 1):
    print(sites.delete_site_by_id(i))
```
