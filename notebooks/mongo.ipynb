```python
from peterbot.data import mongo
from datetime import datetime
```

```python
db = mongo.get_database()
collection = db["test_collection"]

# ---------- WRITE ----------
doc = {
    "name": "Joseph",
    "message": "hello mongo",
    "created_at": datetime.now(),
}

insert_result = collection.insert_one(doc)
print(f"Inserted document ID: {insert_result.inserted_id}")

# ---------- READ ----------
fetched = collection.find_one(
    {"_id": insert_result.inserted_id},
    {"_id": 0}  # hide MongoDB internal ID
)

print("Fetched document:")
print(fetched)
```

# Backups & Exports

### Backup database

```python
path = mongo.backup()
print(f"Database backup created at: {path}")
```

```python
path.name
```

### Restore Database

```python
# Latest backup by default
from pathlib import Path


path = mongo.restore(Path(path))
print(f"Database restored from: {path}")
```

```python
# Latest backup by default
path = mongo.restore()
print(f"Database restored from: {path}")
```

### Export

```python
path = mongo.export()
print(f"Database exported to: {path}")
```
